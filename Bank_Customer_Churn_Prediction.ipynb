### Divide feature in Independent and Dependent variable 

```python
X=final_df.drop(['Exited','Card Type'],axis =1)
y=final_df['Exited']
X.head(2)
```

```python
y.head(2)
```

```python
from imblearn.over_sampling import RandomOverSampler
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,recall_score
from sklearn.model_selection import KFold,GridSearchCV,RandomizedSearchCV
from collections import Counter
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_curve,auc,roc_auc_score
```

### Train and Test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test=train_test_split(X,y, test_size=0.30)
```

```python
X_train
```

#### Feature Selection

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier

feature_select=(RandomForestClassifier())
feature_select.fit(X_train,y_train)
```

```python
plt.figure(figsize=(14,10))
feat_importances=pd.Series(feature_select.feature_importances_, index=X.columns)
feat_importances.nlargest(20).plot(kind="bar")
```

######  Here, the  feature importance is  so high for Complain feature and when the single  feature has such high feature importance  then the model highly depend on that particular  feature to predict,  this situation can leads to  overfitting  condition which is not good  model for prediction because  other feature become less important but which is  important for the prediction.  To prevent that situation we remove Complain feature.

```python
X_train.drop(['Complain'] ,axis=1 ,inplace=True)
X_test.drop(['Complain'] ,axis=1 ,inplace=True)
```

```python
X_train
```

```python
X_test
```

```python
from collections import Counter
os= RandomOverSampler(sampling_strategy=0.80)
X_train_ns,y_train_ns=os.fit_resample(X_train,y_train)
print("the number of classes before fit {}".format(Counter(y_train)))
print("the number of classes after fit {}".format(Counter(y_train_ns)))
```

```python
X_train_ns
```

### Hyperparameter Tunning

```python
log_grid={'C':10.0 **np.arange(-2,3),'penalty':['l1','l2']}


random_grid={'n_estimators':[int(x) for x in np.linspace(start=100, stop=1000, num=10)],
             'max_features':['auto','sqrt'],
             'criterion':['gini','entropy'],
             'max_depth':[2,4,6,10]}

tree_grid={'criterion':['gini','entropy','log_loss'],
           'max_features':['auto','sqrt','log2'],
           'max_depth':[3, 5, 7],
           'min_samples_split': [2, 5, 10],
           'min_samples_leaf': [1, 2, 3]
           
          }

knn_grid={'n_neighbors':[10, 15, 20,],
         'algorithm':['auto','ball_tree', 'kd_tree',],
          'weights': ['uniform', 'distance'],
         }


parameter={'logistic_Regression':log_grid,
            'Decision_tree':tree_grid,
            'RandomForest':random_grid,
            'KNeighbors':knn_grid 
        }
        
        
Classifier={'logistic_Regression':LogisticRegression(),
           'Decision_tree':DecisionTreeClassifier(),
           'RandomForest':RandomForestClassifier(),
           'KNeighbors':KNeighborsClassifier()
        }
```

### Model Train

```python
models_accuracy={}
models_recall={}
best_param={}


plt.figure(figsize=(7,5))

for classifier_name, classifier in Classifier.items():
    params=parameter[classifier_name]
    clf=RandomizedSearchCV(classifier,param_distributions=params,n_iter=7,cv=5)
    clf.fit(X_train_ns,y_train_ns)
    y_pred=clf.predict(X_test)
    y_prob=clf.predict_proba(X_test)[:, 1]
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)
    auc = roc_auc_score(y_test, y_prob)
    print(auc)
    plt.plot(fpr, tpr, label=f'AUC = {auc:.2f}')
    print(accuracy_score(y_test,y_pred))
    print(confusion_matrix(y_test,y_pred))
    print(classification_report(y_test,y_pred))
    models_accuracy.update({classifier_name: accuracy_score(y_test,y_pred)})
    models_recall.update({classifier_name:recall_score(y_test,y_pred)})
    best_param.update({classifier_name:clf.best_params_})
    
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
    
```

```python
model_name=list(models_accuracy.keys())
model_accuracy=list(models_accuracy.values())
plt.figure(figsize=(5,5))
plt.bar(model_name,model_accuracy)
plt.xticks(rotation=30,ha='right')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison of different models')
```

```python
best_param
```

```python
x=0
model_select=""
for i,j in models_accuracy.items():
    if (j>x):
        x=j
        model_select=i
        
model_select
```

```python
best_parameter={'n_estimators':[900],
  'max_features': ['auto'],
  'max_depth': [10],
  'criterion':['gini']
  }

rf_classifier=""
for k in Classifier.items():
    if k[0]==model_select:
        rf_classifier=k[1]
rf_clf=RandomizedSearchCV(rf_classifier,param_distributions=best_parameter,n_iter=7,cv=5)
rf_clf.fit(X_train_ns,y_train_ns)
y_pred=rf_clf.predict(X_test)
print("Accuracy_score:",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
